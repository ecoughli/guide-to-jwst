```python
import os
import webbpsf
import matplotlib.pyplot as plt
# if you are in a folder that has the webbpsf data, you can create a NIRCam instance, 
# select your detector and filter, and generate a psf using the calc_psf command.
os.environ['WEBBPSF_PATH'] = './webbpsf-data'
nrc = webbpsf.NIRCam()
nrc.detector = 'NRCB1'
nrc.filter = 'F115W'
psf = nrc.calc_psf(oversample=4, fov_pixels=11)
webbpsf.display_psf(psf)
```

```python
grid = nrc.psf_grid(num_psfs=16, all_detectors=False)
grid.plot_grid()
grid.plot_grid(deltas=True)
```
